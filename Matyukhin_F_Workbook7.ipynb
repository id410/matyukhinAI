5. –ï—Å–ª–∏ ùë¶ùëë < 0, —Ç–æ —Å–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞—Ç—å –≤–µ—Å–∞ ùë§0 = ùë§0 + ùëéùëë, ùë§ùëñ =

# –ü—Ä–∏–º–µ—Ä 1.1.1

```python
class Perceptron:
  def __init__(self, N):
    self.w = [0 for _ in range(N)]
  
  def calc(self, x):
    res = 0
    for (i, value) in enumerate(self.w):
      res += value * x[i]

    return res

  def sign(self, x):
    return 1 if self.calc(x) > 0 else -1

  def learn(self, la, x, y):
    if y * self.calc(x) > 0:
      return
    
    for (i, value) in enumerate(x):
      self.w[i] += la * y * value
    


  def learning(self, la, T):
    for i in range(100):
      for t in T:
        self.learn(la, t[0], t[1])
```

```python
perceptron = Perceptron(2)
la = .1

T = [
     [[2,1], 1],
     [[3,2], 1],
     [[4,1], 1],
     [[1,2], -1],
     [[2,3], -1],
     [[5,7], -1] 
]

perceptron.learning(la, T)

print(perceptron.w)
```

# –ü—Ä–∏–º–µ—Ä 1.1.2

```python
import numpy as np

def sigmoid(x):
  return 1/(1 + np.exp(-x))

class Neuron:
  def __init__(self, weigths, bias):
    self.weights = weigths
    self.bias = bias

  def feed_forward(self, inputs):
    total = np.dot(self.weights, inputs) + self.bias
    return sigmoid(total)

weights = np.array([0,1])
bias = 4
neuron = Neuron(weights, bias)

inputs = np.array([2,3])
print(neuron.feed_forward(inputs))
```

```python
class NeuralNetwork:
  def __init__(self, x, y):
    self.input = x
    self.weigths1 = np.random.rand(self.input.shape[1], 4)
    self.weigths2 = np.random.rand(4, 1)
    self.y = y
    self.output = np.zeros(y.shape)

  def feed_forward(self):
    self.layer1 = sigmoid(np.dot(self.input, self.weigths1))
    self.output = sigmoid(np.dot(self.input, self.weigths2))

  def backprop(self):
    d_weigths2 = np.dot(self.layer1.T, (2*(self.y - self.output) 
    * sigmoid(self.output)))

    d_weigths1 = np.dot(self.input.T, (np.dot(2*(self.y - self.output)
    * sigmoid(self.output), self.weigths2.T)
    * sigmoid(self.layer1)))

    self.weigths1 += d_weigths1
    self.weigths2 += d_weigths2
```

# –ó–∞–¥–∞–Ω–∏–µ
–†–µ–∞–ª–∏–∑—É–π—Ç–µ –∫–ª–∞—Å—Å—ã –Ω–µ–π—Ä–æ–Ω–Ω—ã—Ö —Å–µ—Ç–µ–π —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –¥—Ä—É–≥–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π
–∞–∫—Ç–∏–≤–∞—Ü–∏–∏



```python


def sigmoid(x):
  return 1/(1 + np.exp(-x))

class Neuron:
  def __init__(self, weights, bias):
    self.weights = weights
    self.bias = bias

  def feed_forward(self, input):
    total = np.dot(input, self.weights) + self.bias
    return sigmoid(total)
```

# –ó–∞–¥–∞–Ω–∏–µ
–†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∫–ª–∞—Å—Å—ã –Ω–µ–π—Ä–æ—Å–µ—Ç–µ–π –ø–æ –∞–Ω–∞–ª–æ–≥–∏–∏ —Å –∫–ª–∞—Å—Å–æ–º OurNeuralNetwork.

‚àí –≤—ã—Ö–æ–¥ (ùëú1).
–ù–µ–π—Ä–æ–Ω—ã –∏–º–µ—é—Ç –∏–¥–µ–Ω—Ç–∏—á–Ω—ã–µ –≤–µ—Å–∞ –∏ –ø–æ—Ä–æ–≥–∏:
‚àí ùë§ = [0.5, 0.5, 0.5]
‚àí ùëè = 0

```python
class NeuralNetwork:
  '''
    - 3 input (x1, x2, x3)
    - 3 hide 1 layer (h1, h2, h3)
    - 1 output (o1)

    for all:
    - w = [0.5, 0.5, 0.5]
    - b = 0
  '''

  def __init__(self):
    weigths = np.array([.5, .5, .5])
    bias = 0

    self.h1 = Neuron(weigths, bias)
    self.h2 = Neuron(weigths, bias)
    self.h3 = Neuron(weigths, bias)

    self.o1 = Neuron(weigths, bias)

  def feed_forward(self, input):
    out_h1 = self.h1.feed_forward(input)
    out_h2 = self.h2.feed_forward(input)
    out_h3 = self.h3.feed_forward(input)

    out_o1 = self.o1.feed_forward(
        np.array([out_h1, out_h2, out_h3])
    )

    return out_o1
```

‚àí ùë§ = [1, 0];
‚àí ùëè = 1.

```python
class NeuralNetwork:
  '''
    - 2 input (x1, x2)
    - 2 hide 1 layer (h1, h2)
    - 2 output (o1, o2)

    for all:
    - w = [1, 0]
    - b = 1
  '''

  def __init__(self):
    weigths = np.array([1, 0])
    bias = 1

    self.h1 = Neuron(weigths, bias)
    self.h2 = Neuron(weigths, bias)

    self.o1 = Neuron(weigths, bias)
    self.o2 = Neuron(weigths, bias)

  def feed_forward(self, input):
    out_h1 = self.h1.feed_forward(input)
    out_h2 = self.h2.feed_forward(input)

    out_o1 = self.o1.feed_forward(
        np.array([out_h1, out_h2])
    )
    
    out_o2 = self.o2.feed_forward(
        np.array([out_h1, out_h2])
    )

    return [out_o1, out_o2]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X_digits,
    Y_digits,
    test_size=.2,
    stratify=Y_digits,
    random_state=123
)
```

```python
from sklearn.neural_network import MLPClassifier

mlp_classifier = MLPClassifier(random_state=123)
mlp_classifier.fit(X_train, Y_train)
```

```python
Y_pred = mlp_classifier.predict(X_test)

print(Y_pred[:15])
print(Y_test[:15])
```

```python
print("Test Accuracy: ", mlp_classifier.score(X_test, Y_test))
print("Training Accuracy: ", mlp_classifier.score(X_train, Y_train))
```

```python
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(Y_test, Y_pred):
  conf_mat = confusion_matrix(Y_test, Y_pred)

  fig = plt.figure(figsize=(6,6))
  plt.matshow(conf_mat, cmap=plt.cm.Blues, fignum=1)
  plt.yticks(range(10), range(10))
  plt.xticks(range(10), range(10))
  plt.colorbar()
  for i in range(10):
    for j in range(10):
      plt.text(i-.2, j+.1, str(conf_mat[j, i]), color='tab:red')
      
plot_confusion_matrix(Y_test, Y_pred)
```
